# Getting the Data

## Codebook Preparation

```python
codebook = pd.read_excel('PIAAC\Cycle 1\International Codebook_PIAAC Public-use File (PUF) Variables and Values.xlsx')
print(codebook.shape)

codebook_vals = pd.read_excel('PIAAC\Cycle 1\International Codebook_PIAAC Public-use File (PUF) Variables and Values.xlsx', sheet_name='Values')
print(codebook_vals.shape)
```

```python
# Additing it to a dictionary
codebook_vals_dict = {}

for index, row in codebook_vals.iterrows():
    variable_name = row['Variable Name']
    coded_value = str(row['Value (SAS)'])
    value_label = row['Value Label']
    
    coded_value = coded_value.lstrip('.') # remove '.' at beginning of some codes

    if variable_name in codebook_vals_dict:
        codebook_vals_dict[variable_name][coded_value] = value_label
    else:
        codebook_vals_dict[variable_name] = {coded_value: value_label}
```

```python
codebook_vals_dict_SPSS = {}

for index, row in codebook_vals.iterrows():
    variable_name = row['Variable Name']
    coded_value = str(row['Value (SPSS)'])
    value_label = row['Value Label']
    
    coded_value = coded_value.lstrip('.') # remove '.' at beginning of some codes

    if variable_name in codebook_vals_dict_SPSS:
        codebook_vals_dict_SPSS[variable_name][coded_value] = value_label
    else:
        codebook_vals_dict_SPSS[variable_name] = {coded_value: value_label}
```

# Cleaning

```python
# Checking null amount
dfs_raw = {'jpn_raw': jpn_raw, 'kor_raw': kor_raw, 'kaz_raw': kaz_raw, 'sgp_raw': sgp_raw}

for df_name, df in dfs_raw.items():
    print('Country ', df_name)
    print('Missing values:')
    print(df.isna().sum())
    print('\n---------------------------------------------------------------------\n')
```

```python
jpn_drop = jpn_raw.copy()
kor_drop = kor_raw.copy()
kaz_drop = kaz_raw.copy()
sgp_drop = sgp_raw.copy()

# values to keep
tokeep = ['SEQID', 'PVPSL1', 'PVPSL2', 'PVPSL3', 'PVPSL4', 'PVPSL5', 'PVPSL6', 'PVPSL7', 'PVPSL8', 'PVPSL9','PVPSL10']

dfs_drop = {'jpn': jpn_drop, 'kor': kor_drop, 'kaz': kaz_drop, 'sgp': sgp_drop}

for df_name, df in dfs_drop.items():
    df.dropna(subset=tokeep, inplace=True)
    df.drop_duplicates(inplace=True)
```

```python
for df_name, df in dfs_drop.items():
    print('Country: ', df_name)
    print('Missing values:')
    print(df.isna().sum())
    print('\n---------------------------------------------------------------------\n')
```

```python
# size difference
for df_name, df in dfs_raw.items():
    print('Country: ', df_name)
    print(df.shape)
    
print('\n---------------------------------------------------------------------\n')

for df_name, df in dfs_drop.items():
    print('Country: ', df_name)
    print(df.shape)
    
# 23463, 1328 - raw
# 16422, 1328 - new
```

```python
jpn_null = jpn_drop.copy()
kor_null = kor_drop.copy()
kaz_null = kaz_drop.copy()
sgp_null = sgp_drop.copy()

dfs_nulls = {'jpn': jpn_null, 'kor': kor_null, 'kaz': kaz_null, 'sgp': sgp_null}

for df_name, df in dfs_nulls.items():
    print('Country: ', df_name)
    print(df.shape)
    null_count = df.isna().sum()
    
    missing_columns = [col for col, count in null_count.items() if count > 0]
    col_all_nan = df.columns[df.isna().all(axis=0)]
    
    if not col_all_nan.empty:
        print('Columns with no values: ', len(col_all_nan))
        print(col_all_nan)
        print()
        
    if missing_columns:
        total_missing_values = null_count[missing_columns].sum()
        print('Total columns with missing values: ', len(missing_columns))
        print('Total missing values: ', total_missing_values)
        
    print('\n---------------------------------------------------------------------\n')
```

```python
# seeing which  cols are always null
consistent_null_cols = set()
all_null_cols_list = []

for df_name, df in dfs_nulls.items():
    col_all_nan = df.columns[df.isna().all(axis=0)]

    if not col_all_nan.empty:
        # Add the set of columns to the list
        all_null_cols_list.append(set(col_all_nan))

# Find the intersection of sets in the list
if all_null_cols_list:
    consistent_null_cols = set.intersection(*all_null_cols_list)

print(len(consistent_null_cols))
print(consistent_null_cols)


# drop the col
for df_name, df in dfs_nulls.items():
    df.drop(columns=consistent_null_cols, inplace=True)


# finding definition in codebook
for col in consistent_null_cols:
    if col in codebook['Name'].values:
        # Retrieve label and domain
        row = codebook[codebook['Name'] == col].iloc[0]
        print(row)
        print('----------------------------------------------------')
    else:
        print('Column Name: ', col, '(No information in codebook)')
        print('----------------------------------------------------')
```

## Combining after cleaning

# Making it more readable

```python
num_type = codebook[codebook['Type'].str.contains('Integer|Numeric/floating point')]
num_labels = num_type['Name'].tolist()
print(len(num_labels))
```

```python
# fix the dtype firsst
df_dtype = df_raw.copy()

for col in num_labels:
    if col in df_dtype.columns:
        df_dtype[col] = pd.to_numeric(df_dtype[col], errors='coerce')

display(df_dtype.dtypes)
```

```python
# fill nulls with 'N'
df_fill = df_dtype.fillna('N')

null_columns = df_fill.columns[df_fill.isnull().sum() > 0].tolist()

# Print the null columns
print(len(null_columns))
```

## Deciphering with the codebook

```python
df_cb = df_fill.astype(str)

for col in df_cb.columns:
    df_cb[col] = df_cb[col].astype(str).apply(lambda x: x.rstrip('0').rstrip('.') if '.' in x else x)
    
display(df_cb.head())
```

```python
for col, mapping in codebook_vals_dict.items():
    if col in df_cb.columns:
        df_cb[col] = df_cb[col].apply(lambda x: mapping.get(x, x))

for col, mapping in codebook_vals_dict_SPSS.items():
    if col in df_cb.columns:
        df_cb[col] = df_cb[col].apply(lambda x: mapping.get(x, x))

df_cb.head()
```

## Fixing the 'N's (null values)

```python
df_ns = df_cb.copy()
```

```python
columns_with_N = df_ns.columns[(df_ns == 'N').any()]

print("Columns with 'N':")
print(columns_with_N)
print(len(columns_with_N))
```

```python
N_dict = {}

for item in columns_with_N:
    if item in codebook_vals_dict:
        for key, value in codebook_vals_dict[item].items():
            if 'Missing' in str(value) or 'Not ' in str(value):
                print(f"Column: {item}, Key: {key}, Value: {value}")
                N_dict[item] = value
                
len(N_dict)
```

```python
for col, new_val in N_dict.items():
    df_ns[col] = df_ns[col].replace('N', new_val)
    
df_ns.replace('N', 'Not Available', inplace=True)
```

```python
df_clean = df_ns.copy()

display(df_clean.head())

df_clean.to_csv('data_cb.csv', index=False)
```
